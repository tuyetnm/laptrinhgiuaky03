```python
#Exploration
#Data Distribution + Sample Selection
#Phân phối dữ liệu và lựa chọn mẫu
import pandas as pd
cat_col = list(df.select_dtypes(include=['object']).columns)
num_col = list(df.select_dtypes(exclude=['object']).columns)
print(f"cat_col: {len(cat_col)}\nnum_col: {len(num_col)}")
#ách cột dữ liệu thành hai danh sách: cat_col và num_col.
```

```python
#trực quan hóa dữ liệu phân loại (categorical data) trong các cột của danh sách cat_col 
#bằng cách sử dụng các biểu đồ cột (bar charts) được sắp xếp thành một lưới 2x3.
import seaborn as sns
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 3, figsize=(10, 8))
fig.subplots_adjust(hspace=0.8, wspace=0.8)

for ax, var in zip(axes.flatten(), cat_col):
    sns.countplot(data=df, x=var, ax=ax, color='black')
    ax.set_title(var)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.show()
```

```python
#đánh giá tương quan giữa các biến số
import numpy as np
import matplotlib.pyplot as plt

corr_matrix = df[num_col].corr()

fig, ax = plt.subplots(figsize=(12, 10))
im = ax.imshow(corr_matrix, cmap='icefire')

ax.set_xticks(np.arange(len(num_col)))
ax.set_yticks(np.arange(len(num_col)))
ax.set_xticklabels(num_col)
ax.set_yticklabels(num_col)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

for i in range(len(num_col)):
    for j in range(len(num_col)):
        text = ax.text(j, i, format(corr_matrix.iloc[i, j], ".2f"),
                       ha="center", va="center", color="black")

plt.colorbar(im)
plt.show()
```
